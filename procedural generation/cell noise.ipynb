```python
import random
import toyplot
```

```python
size = 50

cells = {(i,j):1 for i in range(size) for j in range(size)}

features = random.sample(list(cells),6)
features

def manhattan(a, b):
    xa, ya = a
    xb, yb = b
    return abs(xa-xb)+abs(ya-yb)

def euclidean(a, b):
    xa, ya = a
    xb, yb = b
    return ((xa-xb)**2+(ya-yb)**2)**0.5

def closest(a, bs):
    return min(bs, key=lambda x: euclidean(a,x))

for cell in cells.keys():
    cells[cell] = int(euclidean(cell, closest(cell, features)))

matrix = [[cells[(i,j)] for i in range(size)] for j in range(size)]

toyplot.matrix(matrix)
```
