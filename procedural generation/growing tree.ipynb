```python
import random
```

```python
class Maze():
    UNDECIDED = 1
    BOUNDARY = 2
    CUT = 0
    WALL = 3
    OUTSIDE = -1
    def __init__(self, size):
        self.size = size
        self.cells = [[self.UNDECIDED for _ in range(size)] for _ in range(size)]
        for i in range(size):
            for j in range(size):
                if i==0 or i==size-1 or j==0 or j==-1:
                    self.cells[i][j]=3
        self.set_at((random.randint(1,size-2), 0),self.CUT)
        self.set_at((0,random.randint(1,size-2)),self.CUT) 
        cutting = True
        while cutting:
            cutting = False
            boundary = self.get_if(self.BOUNDARY)
            if boundary:
                cutting = True
                cell = boundary[-1]
                if self.cut_neighbours(cell)==1:
                    self.set_at(cell,self.CUT)
                else:
                    self.set_at(cell,self.WALL)
            cut = self.get_if(self.CUT)
            for cell in cut:
                for neighbour in self.neighbours(cell):
                    if self.get_at(neighbour)==self.UNDECIDED:
                        self.set_at(neighbour, self.BOUNDARY)
                        cutting = True
        
    def neighbours(self, coords, lateral=True, diagonal=False):
        p, q = coords
        result = []
        if lateral:
            for i, j in [(-1,0),(1,0),(0,-1),(0,1)]:
                x, y = p+i, q+j
                if (x>=0)&(x<self.size)&(y>=0)&(y<self.size):
                    result.append((x,y))
        if diagonal:
            for i, j in [(-1,-1),(-1,1),(1,-1),(1,1)]:
                x, y = p+i, q+j
                if (x>=0)&(x<self.size)&(y>=0)&(y<self.size):
                    result.append((x,y))
        return result

    def set_at(self, coords, value):
        p, q = coords
        self.cells[p][q] = value

    def get_at(self, coords):
        p, q = coords
        return self.cells[p][q]  

    def get_if(self, value):
        return [(i, j) for i in range(self.size) for j in range(self.size) if self.get_at((i, j))==value]
    
    def cut_neighbours(self, coords):
        return len([n for n in self.neighbours(coords) if self.get_at(n)==self.CUT])
```
