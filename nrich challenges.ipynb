[Euclid's algorithm](https://nrich.maths.org/5928)

```python
def f(x, y):
    if x==y:
        return x
    elif x>y:
        return f(x-y, y)
    else:
        return f(x, y-x)
```

```python
f(2*3*5*7,3*5*7*11)
```

[Building $\sin x$](https://nrich.maths.org/5622)

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
fig, ax = plt.subplots()
ax.set_ylim(bottom=-3,top=3)
xx = np.linspace(-2*np.pi, 2*np.pi, 50)
yy = np.sin(xx)
y1 = xx-(xx**3)/6+(xx**5)/120-(xx**7)/math.factorial(7)+(xx**9)/math.factorial(9)-(xx**11)/math.factorial(11)
ax.plot(xx,yy,)
ax.plot(xx,y1)
```

```python
def fake_sin(x, terms=5):
    return sum([(-1)**i*x**(2*i+1)/math.factorial(2*i+1) for i in range(terms)])
```

```python
fig, ax = plt.subplots()
ax.set_ylim(bottom=-2,top=2)
xx = np.linspace(-3*np.pi, 3*np.pi, 50)
yy = np.sin(xx)
y1 = fake_sin(xx, terms=10)
ax.plot(xx,yy,)
ax.plot(xx,y1)
```
