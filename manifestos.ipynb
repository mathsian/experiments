```python
import pandas as pd
import toyplot
import matplotlib.pyplot as plt
%matplotlib notebook
plt.style.use('ggplot')
```

```python
df = pd.read_csv("MPDataset_MPDS2016b.csv")
df.head(3)
```

```python
df[(df.country==51)&(df.partyabbrev=="Labour")].rile.head(3)
```

```python
df["dt"] = pd.to_datetime(df.edate,utc=True,format="%d/%m/%Y")
```

```python
df.dt
```

```python
labour = df[(df.country==51)&(df.partyabbrev=="Labour")]
conservative = df[(df.country==51)&(df.partyabbrev=="Conservatives")]
```

```python
list(conservative.columns)
```

```python
plt.plot(conservative.dt,conservative.rile,label="CON LR",color="blue")
plt.plot(conservative.dt,conservative.pervote,label="CON VOTE",color="blue")
plt.plot(labour.dt,labour.rile,label="LAB LR",color="red")
plt.plot(labour.dt,labour.pervote,label="LAB VOTE",color="red")
plt.legend()
```

```python
canvas, axes, marks = toyplot.plot(labour.dt.astype('int')//1000,labour.rile,color="red")
axes.plot(conservative.dt.astype('int')//1000,conservative.rile,color="blue")
axes.x.ticks.locator = toyplot.locator.Timestamp()
```
